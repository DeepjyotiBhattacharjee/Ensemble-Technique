```python
import numpy as np
import pandas as pd
import os
import sys
```

```python
from sklearn import datasets
data = datasets.load_iris()
data
```

```python
data.data
```

```python
X = pd.DataFrame(data.data,columns=['sepal length','sepal width','petal length','petal width'])
X
```

```python
data.target
```

```python
y = pd.DataFrame(data.target,columns=['species'])
y
```

```python
data = pd.concat([X,y],axis=1)
data
```

```python
import seaborn as sns
```

```python
sns.pairplot(data,hue='species')
```

```python
data[data['species']!=0]
```

```python
optimised_data = data[data['species']!=0][['sepal length','sepal width','species']]
optimised_data
```

```python
sns.pairplot(optimised_data,hue='species')
```

```python
# we will use this optimised data to build our classifier
```

```python
optimised_data.shape
```

```python
X = optimised_data.iloc[:,0:-1]
y = optimised_data.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
# Since voting is heterogeneous, let's import 3 different models
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clf1 = LogisticRegression()
clf2 = KNeighborsClassifier()
clf3 = DecisionTreeClassifier()
```

```python
estimators = [('lr',clf1),('knn',clf2),('dtc',clf3)]
```

```python
estimators
```

```python
estimators[0]
```

```python
estimators[1]
```

```python
estimators[2]
```

```python
for estimator in estimators:
    print(estimator[1])
```

```python
for estimator in estimators:
    x = cross_val_score(estimator[1],X,y,cv=10,scoring='accuracy')
    print(estimator[0],np.round(np.mean(x),2))
```

```python
# these are individual results from each model
```

# voting classifier

### hard voting

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators=estimators,voting='hard')
```

```python
vc
```

```python
vc_accuracy = cross_val_score(vc,X,y,cv=10,scoring='accuracy') # 10 accuracies since cv=10
vc_accuracy
```

```python
# let's calculate the mean accuracy/
np.round(np.mean(vc_accuracy),2)
```

### soft voting

```python
vc_s = VotingClassifier(estimators=estimators,voting='soft')
```

```python
vc_s
```

```python
vc_accuracy = cross_val_score(vc_s,X,y,cv=10,scoring='accuracy') # 10 accuracies since cv=10
vc_accuracy
```

```python
# let's calculate the mean accuracy/
np.round(np.mean(vc_accuracy),2)
```
