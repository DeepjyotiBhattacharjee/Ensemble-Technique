```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
df
```

```python
np.random.seed(42)
```

```python
X = np.random.rand(100,1)-0.5
X
```

```python
y = 3*X[:,0]**2 + 0.05*np.random.randn(100)
```

```python
y
```

```python
df['X'] = X.reshape(100)
```

```python
df
```

```python
df['y'] = y
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
```

```python
df['pred1 (base model)'] = df['y'].mean()
```

```python
df
```

```python
df['res1'] = df['y'] - df['pred1 (base model)']
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.plot(df['X'],df['pred1 (base model)'],color='green')
```

```python
# we are going to train our first DT on top of X, res1
```

```python
X = df['X'].values.reshape(100,1)
X
```

```python
y = df['res1']
y
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1 = DecisionTreeRegressor(max_leaf_nodes=8)
dt1.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15,15))
plot_tree(dt1)
plt.show()
```

```python
df['pred2'] = dt1.predict(X)
```

```python
df
```

```python
# loss (res2) = y-(base_model + alpha*pred2)
```

```python
BASE = df['pred1 (base model)']
BASE
```

```python
alpha = 0.1
```

```python
df["res2"] = df['y'] - (BASE + alpha*df['pred2'])
```

```python
df
```

```python
X =df['X']
X
```

```python
y = df['res2']
y
```

```python
X = X.values.reshape(100,1)
y = y.values
```

```python
X
```

```python
y
```

```python
dt2 = DecisionTreeRegressor(max_leaf_nodes=8)
dt2.fit(X,y)
```

```python
dt2.predict(X)
```

```python
plt.figure(figsize=(15,15))
plot_tree(dt2)
plt.show()
```

```python
query = [[0.45071431]]
```

```python
pred1 = dt1.predict(query)
pred1
```

```python
pred2 = dt2.predict(query)
pred2
```

```python
final_pred = BASE[0] + alpha*pred1 + alpha*pred2
```
